## Hyperparameters
Besides weight parameters, there are various other degrees of freedom that determine the performance of the neural network:
* the number of hidden layers
* the number of neurons per layer
* the kind of layer (dense, convolution, dropout, recurrent)
* the learning rate
* the activation function
* the optimizer
and various other. 

These *extra* parameters are also called **hyperparameters**. Choosing the *right* or optimal (for a given task) hyperparameters
is where domain knowledge comes into the game. E.g. in computer vision classifications tasks, convolutional neural networks perform best, because pattern reckognition relies to a large degree on filtering functions and these filters can be represented by convolution operators that act on the signal tensors that propagate through the neural network. Similarly, different activation functions are to be preferred for different tasks, e.g. for binary or multiple classification tasks vs. regression tasks.

Here, we pretend we have no such domain expertise and try to optimize the hyperparameters using a **genetic algorithm**.

## Genetic Algorithm for hyperparameter tuning

* **Genes** on a NN chromosome encode the NN hyperparameters
* **Developmental phase**: Maturing of the phenotype (= NN training) within the boundaries set by the genotype
* **Competition phase**: Evaluate the NN on validation data, performance score = $\frac{\text{number of correct labels}}{\text{number of all labels}}$
* **Generation turnover phase**: Selection, reproduction, mutation
